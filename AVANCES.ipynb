### C. Cantidad y porcentaje de valores faltantes por variable.

```python
faltantes = {'Cantidad': [],
             'Porcentaje %': []}

for x in df.columns:
    cantidad = int(df[x].isna().sum())
    porcentaje = cantidad*100/n_registros

    faltantes['Cantidad'].append(cantidad)
    faltantes['Porcentaje %'].append(porcentaje)

faltantes_df = pd.DataFrame(faltantes, index= df.columns)
faltantes_df
```

### D. Cantidad de valores únicos.

```python
unicos = {'Cantidad': []}

for x in df.columns:
    cantidad = int(df[x].nunique())
    unicos['Cantidad'].append(cantidad)

unicos_df = pd.DataFrame(unicos, index= df.columns)
unicos_df
```
