### C. Cantidad y porcentaje de valores faltantes por variable.

```python
faltantes = {'Cantidad': [],
             'Porcentaje %': []}

for x in df.columns:
    cantidad = int(df[x].isna().sum())
    porcentaje = cantidad*100/n_registros

    faltantes['Cantidad'].append(cantidad)
    faltantes['Porcentaje %'].append(porcentaje)

faltantes_df = pd.DataFrame(faltantes, index= df.columns)
faltantes_df
```

#### Revisión de que registros quedaron con valores faltantes

```python
faltantes = {'Cantidad': [],
             'Porcentaje %': []}

for x in df.columns:
    cantidad = int(df[x].isna().sum())
    porcentaje = cantidad*100/n_registros

    faltantes['Cantidad'].append(cantidad)
    faltantes['Porcentaje %'].append(porcentaje)

faltantes_df = pd.DataFrame(faltantes, index= df.columns)
faltantes_df[faltantes_df['Cantidad'] > 0]
```

### I. Identificación de problemas potenciales de calidad de datos.

- Todas las columnas tienen valores nulos.
- Hay columnas en las que los datos tienen longitudes menores o mayores a las esperadas, pudiendo reflejar falta de información o que se tiene más datos que un solo elemento.

### Plan de limpieza.

##### CODIGO

No hay inconsistencias con el código.

##### DISTRITO

```python
for i, x in df["DISTRITO"].items():

    if pd.isna(x):
        df.at[i, "DISTRITO"] = "NA"
```

##### DEPARTAMENTO

```python
df["DEPARTAMENTO"] = (
    df["DEPARTAMENTO"]
      .ffill()
      .bfill()
)
```

Se han rellenado las casillas con valores NaN de departamento con los valores de las casillas inferiores o superiores. Esto se hizo considerando que la lista está en orden alfabetico respecto a estos departamentos.

##### MUNICIPIO

```python
df["MUNICIPIO"] = (
    df["MUNICIPIO"]
      .ffill()
      .bfill()
)
```

Dado que los municipios de cada departamento están escritos en orden alfabético, se establece que los valores NaN sean sustituidos con los valores de las casillas superiores o inferiores, de manera similar a como se hizo con los departamentos.

##### TELEFONO

```python
for i, x in df["TELEFONO"].items():

    if pd.isna(x):
        df.at[i, "TELEFONO"] = "NUMERO_NULO"

    elif len(str(x)) != 8:
        df.at[i, "TELEFONO"] = "NUMERO_INVALIDO"
    
```

```python
df['TELEFONO']
```

Se ha sustituido los valores de las casillas con números telefónicos con cantidad de dígitos distinta de 8 con "NUMERO_INVALIDO". Aquellas casillas con valores NaN, se han sustituido con "NUMERO_NULO".

##### SUPERVISOR

```python
for i, x in df["SUPERVISOR"].items():

    if pd.isna(x):
        df.at[i, "SUPERVISOR"] = "NO ESPECIFICADO"
```

Si no se especifica el supervisor se sustituye por "NO_ESPECIFICADO".

##### DIRECTOR

```python
for i, x in df["DIRECTOR"].items():

    if pd.isna(x):
        df.at[i, "DIRECTOR"] = "NO_ESPECIFICADO"
```

Si no se especifica el director se sustituye por "NO_ESPECIFICADO".

##### NIVEL

```python
for i, x in df["NIVEL"].items():

    if pd.isna(x):
        df.at[i, "NIVEL"] = "DIVERSIFICADO"
```

Se han reemplazado las casillas con valores NaN con "DIVERSIFICADO", ya que el conjunto de datos extraido corresponde a aquel con instituciones de nivel máximo de diversificado.
