##### Importación de librerías.

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
```

##### Dirección de los datos.

```python
url = 'https://www.mineduc.gob.gt/BUSCAESTABLECIMIENTO_GE/wbfBuscar.aspx'
```

##### Captura de datos.

```python
# Listado de departamentos.
departamentos = ['ALTA VERAPAZ', 'BAJA VERAPAZ',
                 'CHIMALTENANGO', 'CHIQUIMULA',
                 'EL PROGRESO', 'ESCUINTLA',
                 'GUATEMALA', 'HUEHUETENANGO',
                 'IZABAL', 'JALAPA',
                 'JUTIAPA', 'PETEN',
                 'QUETZALTENANGO', 'QUICHE',
                 'RETALHULEU', 'SACATEPEQUEZ',
                 'SAN MARCOS', 'SANTA ROSA',
                 'SOLOLA', 'SUCHITEPEQUEZ',
                 'TOTONICAPAN', 'ZACAPA']
```

```python
departamentos_df = []
for dep in departamentos:

    # Abrir Chrome
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(2)

    # Menús desplegables.
    selects = driver.find_elements(By.TAG_NAME, "select")
    Select(selects[0]).select_by_visible_text(dep)
    time.sleep(1)
    selects = driver.find_elements(By.TAG_NAME, "select")
    Select(selects[2]).select_by_visible_text("DIVERSIFICADO")

    # Botón de consulta.
    boton = driver.find_element(By.ID, "_ctl0_ContentPlaceHolder1_IbtnConsultar")
    boton.click()
    time.sleep(3)
    
    # Extraer las tablas de la página
    tablas = pd.read_html(driver.page_source)
    df = tablas[-1]

    departamentos_df.append(df)
    driver.quit()
    time.sleep(3)
```

##### Verificando los resultados.

```python
departamentos_df[0].head()
```

```python
for i in range(0, 21):
    print(departamentos_df[i].iloc[0, 2:3], departamentos_df[i].shape[1])
```

Todos los df tienen los títulos en la primera fila, por lo cual se hará y se eliminará la fila 0, la cual no contiene información.

##### Formateando los DataFrames.

```python
for i, df in enumerate(departamentos_df):
    df = df.drop(columns=[0])
    df.columns = df.iloc[0]
    df = df.iloc[1:]
    departamentos_df[i] = df
```

##### Concatenando todos los Data Frame.

```python
df_final = pd.concat(departamentos_df, axis=0, ignore_index=True)
```

##### Exportando a csv.

```python
df_final.to_csv('DATOS_CRUDOS.csv', index=False)
```
