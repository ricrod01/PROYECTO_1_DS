```python
import pandas as pd
import numpy as np
```

```python
filename= 'DATOS_CRUDOS.csv'
data= pd.read_csv(filename, index_col=0)
df=pd.DataFrame(data)
df.head()
```

Análsis de las variables

```python
df.info()
```

Se observa que todas las variables son cadenas. Son aproximadamente 16 variables. No todas son indispensables, por lo que se proccederá al su correspondiente eliminación del conjunto. 

```python
df=df.drop(columns= ['DISTRITO', 'TELEFONO', 'DIRECCION'])
df
```

```python
df.shape
```

Conteo de valores nulos por variable

```python
# Count missing values in the DataFrame
missing_values = df.isnull().sum()
print(missing_values)
```

```python
di=pd.DataFrame(missing_values)
print("Valores NaN: ",di.sum())
print("Proporción de NaN respecto al total: ", di.sum()/ 9728)
print("Porcentaje de NaN respecto al total: ", (di.sum()/ 9728)*100)
```

```python
df=df.dropna()#eliminación de datos nulos NaN
df
```

```python
print("Dimensiones del dataframe: ",df.shape) #tamaño del dataframe
```

```python
df.info()
```
