### Importación de librerías.

```python
import pandas as pd
import numpy as np
```

### Importación de datos.

```python
df = pd.read_csv('DATOS_LIMPIOS.csv')
```

### Valores duplicados.

```python
print('Hay en total:', df.duplicated().sum(), 'datos duplicados.')
```

### Espacios en blanco.

```python
k = 0
for i in df.columns:
    for j in df[i]:
        if (str(j)[0] == ' ' or str(j)[-1] == ' '):
            print(i)
            k += 1
            break

if k == 0:
    print('No hay espacios ni al inicio ni al final de cada elemento.')
```

### Formato de los teléfonos.

```python
contador = 0
for x in df['TELEFONO']:
    for y in str(x).split('|'):
        if len(y) != 8 and y != 'NO_VALIDO':
            contador += 1
            print('FORMATO NO VALIDO', y)
if contador == 0:
    print('Todos los formatos son válidos.')
```

### Verificando Mayúsculas.

```python
contador = 0
for x in df.columns:
    if (df[x] != df[x].str.upper()).any():
        contador += 1
        print('La variable:', x, 'tiene valores en minúsculas.')
if contador == 0:
    print('Todas las variables están en mayúsculas.')
```

### Verificando tildes.

```python
tildes = ['Á', 'É', 'Í', 'Ó', 'Ú']
contador = 0
for x in df.columns:
    for y in tildes:
        for z in df[x]:
            if y in z:
                contador += 1
                print('La variable:', x, 'tiene valores de', y, '.')
if contador == 0:
    print('Todas las variables no tienen tildes.')
```

### Tipo de dato esperado.

```python
contador = 0
for x in df.columns:
    if df[x].dtype != 'object':
        print(df[x].dtype)
        contador += 1
if contador == 0:
    print('Todas las variables tienen el tipo esperado.')
```

### Verificando nulos.

```python
n_registros, _ = df.shape
faltantes = {'Cantidad': [],
             'Porcentaje %': []}

for x in df.columns:
    cantidad = int(df[x].isna().sum())
    porcentaje = cantidad*100/n_registros

    faltantes['Cantidad'].append(cantidad)
    faltantes['Porcentaje %'].append(porcentaje)

faltantes_df = pd.DataFrame(faltantes, index= df.columns)
faltantes_df[faltantes_df['Cantidad'] > 0]
```

Ya no hay columnas con datos nulos.
